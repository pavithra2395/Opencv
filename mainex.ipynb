```python
#The paddleocr whl package will automatically download the ppocr lightweight model as the default model, which can be customized and replaced according to the section 3 Custom Model.
#detection angle classification and recognition


#Output will be a list, each item contains bounding box, text and recognition confidence

import os
import xml.etree.ElementTree as ET
import glob
import cv2
from paddleocr import PaddleOCR,draw_ocr



path =  r"/home/ppavithra/Desktop/notebook/Opencv-main/facility-onboarding_classification/prediction_images"


files = glob.glob(os.path.join(path,'*.xml'))


# class_name = 'Alphabet_Box1'

ocr = PaddleOCR(use_angle_cls=True, lang='en')

for file in files:

    tree = ET.parse(file)
    root = tree.getroot()
    for object in root.findall('object'):
        name = object.find('name').text
        bnd_box = object.find('bndbox')
        xmin = bnd_box.find('xmin').text
        ymin = bnd_box.find('ymin').text
        xmax = bnd_box.find('xmax').text
        ymax = bnd_box.find('ymax').text


#         print(int(ymin),int(ymax),int(xmin),int(xmax))

        image = file.replace('.xml','.png')
        img = cv2.imread(image)
#         print(img.shape)
#         print(type(img))
#         cv2.imshow('image',img)       
        crop_img = img[int(ymin):int(ymax),int(xmin):int(xmax)]
        result = ocr.ocr(crop_img, cls=True)
        print(result)


        import bson
        cv2.imwrite(f'/home/ppavithra/Desktop/notebook/Opencv-main/facility-onboarding_classification/one-equipment_crops/ {str(bson.ObjectId())}.png ',crop_img)

        print(crop_img)
#         cv2.imshow('crop_img',crop_img)
#         cv2.waitkey(0)
```
