```python
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('6332fd3b7312f6a09ab66252.png')
plt.figure(figsize = (10,6))
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# cv2.imshow('image',gray)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
edges = cv2.Canny(img,100,200)
plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
cv2.imwrite('gray.jpg', edges)
```

```python
from paddleocr import PaddleOCR
ocr = PaddleOCR(use_angle_cls=True, lang='en')
img_path = 'erosionimage.png'
result = ocr.ocr(img_path, det=False, cls=True)
for line in result:
    print(line)
```

```python
import cv2
import numpy as np
 
img = cv2.imread('6332fd397312f6a09ab6624b.png', 0)
kernel = np.ones((5, 5), np.uint8)
img_erosion = cv2.erode(img, kernel, iterations=1)
img_dilation = cv2.dilate(img, kernel, iterations=1)
cv2.imwrite('erosionimage2.png', img_erosion)
cv2.imwrite('dilationimage2.png', img_dilation)
 
cv2.imshow('Input', img)
cv2.imshow('Erosion', img_erosion)
cv2.imshow('Dilation', img_dilation)
 
cv2.waitKey(0)
```
